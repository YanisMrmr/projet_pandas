# Iris
Iris is a famous dataset to test the different representation tools pandas offer with matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
# The simplest thing we can do, is to plot our dataframe
df.plot()
```

```python
#We can also visualize the means the and the quartiles for each variable
df.boxplot()
```

```python
# One boxplot per species of flower

df.boxplot(['PetalLength', 'PetalWidth'], by='Name')
plt.tight_layout() # le padding
```

```python
# All the species of flower in the dataframe
df.Name.unique()
```

```python
# The data type is wrong. Category data type is better when it's always the same string that repeats itself.
df.Name = df.Name.astype('category')
df.Name.unique()
```

```python
# Each category is represented by an int 
# Iris-setosa -> 0, Iris-versicolor -> 1, Iris-virginica ->2
df['Name'].cat.codes
```

```python
# Let's put those codes in another column
df['Name-code'] = df['Name'].astype('category').cat.codes
```

```python
# Creation of a scatter plot with a different color per category
df.plot.scatter(x='SepalLength', y='SepalWidth', c='Name-code', cmap='viridis');
```

```python
df.plot.scatter(x='PetalLength', y='PetalWidth', c='Name-code', cmap='viridis');
```

We found a representation that would perfectly allow the prediction of an unknown flower's name from the width and the length of its petals.
