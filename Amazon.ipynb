# Amazon
We have the amazon stock dataframe.
We will plot the highest price as a function of the time and replot it to simplify it.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("amazon.csv", skiprows = 3) #because the first three lines of the file or comments.
```

```python
df.Date.dtype
# Wrong data type, it would be more adapted in datetime.
```

```python
df.Date = pd.to_datetime(df.Date)
str(df.Date.dtype)
```

```python
df.dropna(subset=['Date'], inplace=True) # We drop the lines with no date
df.set_index('Date', inplace=True) # We set Date as the index of the dataframe
df.head(2)
```

```python
# We want to plot the variable 'High' as a function of the time between
# the July 2018 until the end of 2019
df = df['High'].loc['07-2018':'2019']
```

```python
df.plot()
```

```python
# The graph is very complicated, we can simplfy it by plotting for each week the mean of the values of the week.
# It can be done with the function resample
# Theorically, it would divide the number of points per 7, but this is not totally the case because we have not always following days in the dataframe.
df.resample(rule = pd.Timedelta(unit = 'D', value = 7)).mean().plot()
```
